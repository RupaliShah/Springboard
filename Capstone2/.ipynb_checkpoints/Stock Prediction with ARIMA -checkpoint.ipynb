### Pre-prediction munging and stationarity
Stationary time series has a Constant mean, variance, and auto correlation. i.e. value and predictibility are not drifting over time, and auto correlation structure is not changing over time. Time Series behaves in a certain way and will always behave thay way.

```python
# Check if the Time Series is stationary using the Augmented Dickey Fuller Test(adfuller()). 
# The null hypothesis of the ADF test is that the Times Series not stationary.

print('Augmented Dickey Fuller test results:\n')

adf_test = adfuller(stock_data.Adj_Close, autolag='AIC')
adf_result = pd.Series(adf_test[0:4], index=['Test Statistic', 'p-value', 'Number of lags', 'Number of observations'])
for key, value in adf_test[4].items():
        adf_result['Critical Value (%s)' %key] = value
        
print(adf_result)
```

Critical value is less than than the Test Statistic. p-value is very high (>0.5). So, we accept the Null Hypotheseis that the Time Series is not stationary. 

```python
# Difference the time Series to make it Staionary

log_Adj_Close = np.log(stock_data.Adj_Close)
log_Adj_Close_diff = log_Adj_Close - log_Adj_Close.shift()

plt.figure(figsize=(12,6))
plt.plot(log_Adj_Close_diff)
plt.title('Stationary Time Series of Chevron Stock Prices')
plt.xlabel('Date')
plt.ylabel('Adjusted Closing Price')
plt.show()
```

```python
# The above Time Series looks stationary, Check stationary of the new Time Series using the adfuller()
# The null hypothesis of the ADF test is that the Times Serie snot stationary.

log_Adj_Close_diff.dropna(inplace=True)

adf_test_diff = adfuller(log_Adj_Close_diff, autolag='AIC')
adf_result_diff = pd.Series(adf_test_diff[0:4], index=['Test Statistic', 'p-value', 'Number of lags', 'Number of observations'])
for key, value in adf_test_diff[4].items():
        adf_result_diff['Critical Value (%s)' %key] = value
        
print(adf_result_diff)
#print('ADF Statistic:', result[0])
#print('p-value:', result[1])
```

Since Critical Value is than the Test Statistic, and p-value is less than the significance level (0.05), we reject the null hypotheseis that the Time Series is not stationary. Hence, no more differencing is needed.

###  Find AR and MA values

Build ARIMA(Auto Regressive Integrated Moving Average) model

arima = error + moving averages going back in time 
linear regression with a few times
1. The number of Auto-Regressive Terms (p)
2. The number of differences taken (d)
3. The number of Moving Average Terms (q)

Autocorrelation refers to how correlated a time series is with its past values whereas the ACF is the plot used to see the 
correlation between the points, up to and including the lag unit.


```python
log_Adj_Close_diff.dropna(inplace=True)
lag_acf = acf(log_Adj_Close_diff.values, nlags = 20)
lag_pacf = pacf(log_Adj_Close_diff.values, nlags = 20, method='ols')
```

```python
# Plot acf graph to determine the value of q (Moving Average term)
plt.bar(x = range(len(lag_acf)), height = lag_acf)
plt.axhline(y=0,linestyle='--')
plt.axhline(y=-1.96/np.sqrt(len(log_Adj_Close_diff )),linestyle='--')
plt.axhline(y=1.96/np.sqrt(len(log_Adj_Close_diff)),linestyle='--')
plt.title('ACF Plot of Chevron Stock Prices')
plt.xlabel('Lags')
plt.ylabel('Autocorrelation Coefficients')
plt.show()

plt.plot(lag_acf)
plt.axhline(y=0,linestyle='--')
plt.axhline(y=-1.96/np.sqrt(len(log_Adj_Close_diff )),linestyle='--')
plt.axhline(y=1.96/np.sqrt(len(log_Adj_Close_diff)),linestyle='--')
plt.title('ACF Plot of Chevron Stock Prices')
plt.xlabel('Lags')
plt.ylabel('Autocorrelation Coefficients')
plt.show()
```

```python
# Plot pacf graph to determine the value of p (Auto Regressive term)
plt.bar(x = range(len(lag_pacf)), height = lag_pacf)
plt.axhline(y=0,linestyle='--')
plt.axhline(y=-1.96/np.sqrt(len(log_Adj_Close_diff)),linestyle='--')
plt.axhline(y=1.96/np.sqrt(len(log_Adj_Close_diff)),linestyle='--')
plt.title('PACF Plot of Chevron Stock Prices')
plt.xlabel('Lags')
plt.ylabel('Partial Autocorrelation Coefficients')
plt.show()

plt.plot(lag_pacf)
plt.axhline(y=0,linestyle='--')
plt.axhline(y=-1.96/np.sqrt(len(log_Adj_Close_diff )),linestyle='--')
plt.axhline(y=1.96/np.sqrt(len(log_Adj_Close_diff)),linestyle='--')
plt.title('PACF Plot of Chevron Stock Prices')
plt.xlabel('Lag')
plt.ylabel('Partial Autocorrelation Coefficients')
plt.show()
```

### Build ARIMA model

```python
# (p, d, q) 
# p is the auto regressive term, d is the differencing, q is the moving average term
# disp - int, optional
# If True, convergence information is printed. For the default l_bfgs_b solver, disp controls the frequency of the output 
# during the iterations. disp < 0 means no output in this case.

model = ARIMA(log_Adj_Close, order=(0, 1, 0))  
fitted_model = model.fit(disp=0)  
print(fitted_model.summary())
```

```python
f =fitted_model.forecast(steps=1154)[0]
print(f)
fitted_model.plot_predict(1, 1154)
```

### Build model using auto_arima

```python
# split data into train and test 
# cannot do slice indexing on classs float, so converted to int

data = stock_data.Adj_Close
y_train = data[:int(0.8*(len(data)))]
y_test = data[int(0.8*(len(data))):]
```

```python
# Determine the value of d
# use the KPSS test and ADF test, select the maximum value between the two to be conservative. 

kpss_diffs = ndiffs(y_train.values, alpha=0.05, test='kpss', max_d=6)
adf_diffs = ndiffs(y_train.values, alpha=0.05, test='adf', max_d=6)
n_diffs = max(adf_diffs, kpss_diffs)

print(f"Estimated differencing term: {n_diffs}")
```

```python
# Fit the model

model = auto_arima(y_train,  seasonal=False, stepwise=True,
                     suppress_warnings=True, error_action="ignore",trace=True) 
                    # d=n_diffs, max_p=3, max_order=None)
model.fit(y_train)
```

```python
# Make predictions
 
forecast, conf = model.predict(n_periods=len(y_test), return_conf_int=True)
forecast_df = pd.DataFrame(forecast, index = y_test.index, columns=['Prediction'])
#conf_df = pd.DataFrame(conf, index = y_test.index, columns=['cf'])

#plot the predictions for test set
plt.figure(figsize= (12,6))
plt.plot(y_train, label='Train')
plt.plot(y_test, label='Test')
plt.plot(forecast_df, label='Forecast')
plt.legend(loc='lower right')
plt.title('Chevron Stock Prices Forecast')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.show()
```

```python
#calculate rmse

rmse = sqrt(mean_squared_error(y_test,forecast))
print('RMSE:' , round(rmse,2))
```
