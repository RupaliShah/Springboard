```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
train_df = pd.read_csv('data/train_clean.csv')
test_df = pd.read_csv('data/test_clean.csv')
```

```python
train_df.head()
```

```python
test_df.shape
```

```python
# Average Sales

mean_sales = train_df.Item_Outlet_Sales.mean()
round(mean_sales)
```

```python
# Average sales by product

mean_item_sales = train_df.groupby('Item_Type')['Item_Outlet_Sales'].mean()
round(mean_item_sales)
```

```python
# Average and median sales by product category

mean_category_sales = train_df.groupby('Item_Category')['Item_Outlet_Sales'].aggregate(['mean', 'median'])
round(mean_category_sales)
```

```python
# Average sales by product per Outlet

product_outlet_sales = train_df.groupby(['Outlet_Type', 'Item_Category'])['Item_Outlet_Sales'].mean()
round(product_outlet_sales)
```

```python
product_outlet_sales.unstack(level = -1)
```

```python
_ = sns.distplot(train_df.Item_Visibility)
plt.show()
```

```python
# Item visibility versus Sales - line of best fit
a, b = np.polyfit(train_df.Item_Visibility, train_df.Item_Outlet_Sales, 1)
x_lr = np.array([0.00, 0.35])
y_lr = a * x_lr + b

_ = plt.scatter(train_df.Item_Visibility, train_df.Item_Outlet_Sales, s=None, facecolors = 'lightgreen' )
_ = plt.plot(x_lr, y_lr)

_ = plt.xlabel('Item Visibility')
_ = plt.ylabel('Sales')
_ = plt.title('Item Visibility versus Sales')
plt.show()
```

There is a strong negative linear correlation between item visibility and Sales.

```python
# Pearson Correlation coefficient

corr_vis = np.corrcoef(train_df.Item_Visibility, train_df.Item_Outlet_Sales )
print(corr_vis[0,1]) 
```

The observed significance value of -0.12 shows a negative corrleation between item visibility and sales

```python
corr_mrp = np.corrcoef(train_df.Item_MRP, train_df.Item_Outlet_Sales)
print(corr_mrp[0,1]) 
```

There is a positive correlation between price and sales

```python
# Correlation_matrix

corr = train_df.corr()  
#cmap = sns.light_palette('green') 
cmap = sns.diverging_palette(220, 10, as_cmap=True)
 
sns.heatmap(corr, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.show()
```
